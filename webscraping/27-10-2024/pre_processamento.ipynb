```python
import json
import re
import pandas as pd
```

```python
# 1. Carregar o arquivo JSON bruto
with open("conteudo_paginas_sem_links.json", "r", encoding="utf-8") as f:
    dados = json.load(f)

print(f"Total de páginas carregadas: {len(dados)}")

# 2. Frases para remover
PADROES_REMOVER = [
    r"\nO que você está procurando?",
    r"\nO que você está procurando?\n",
    r"\nNossos produtos\n",
    r"\nSobre nós\n",
    r"\nFonte legivel\n",
    r"\nCategorias\n",
    r"\n? Categorias ?\n",
    r"\nDisponível no Google Play\n"
]

# Compila regex para performance
regex_remocao = re.compile("|".join(PADROES_REMOVER), flags=re.IGNORECASE)

# 3. Função de limpeza
def limpar_texto(texto):
    # Remove padrões exatos
    texto_limpo = regex_remocao.sub("\n", texto)

    # Remove múltiplos \n seguidos (deixa só um)
    texto_limpo = re.sub(r"\n+", "\n", texto_limpo)

    # Remove espaços extras
    return texto_limpo.strip()

# 4. Aplica limpeza em todos os conteúdos
dados_tratados = []
for item in dados:
    url = item["url"]
    conteudo = item["conteudo"]
    conteudo_limpo = limpar_texto(conteudo)
    dados_tratados.append({"url": url, "conteudo": conteudo_limpo})

# 5. Salvar em novo JSON
with open("conteudo_paginas_tratado.json", "w", encoding="utf-8") as f:
    json.dump(dados_tratados, f, ensure_ascii=False, indent=4)

print("Arquivo tratado salvo em conteudo_paginas_tratado.json")
```
